# Ice Breaker Dataset Visualization

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# dataframe - df

df = pd.read_csv('https://raw.githubusercontent.com/rugvedmhatre/NYU-ML-2024-Session-1/main/day2/Ice%20Breaker%20Spreadsheet.csv') # add the link of the dataset file from the GitHub repository
```

## Plot the IMDB Ratings for all the movies
- Use the [plt.scatter() function](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.scatter.html)
- x-axis indicates the name of the movies, y-axis shows the score.
- do not forget to put on the labels for the axes.

```python
names = df['Movie'].values
rating = df['IMDB Rating'].values
```

```python
## To-do
plt.figure(figsize=(15, 5))
plt.scatter(names, rating)
plt.xticks(rotation=90)
plt.show()
```

## Plot the number of movies for each category
- Use the [plt.bar() function](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.bar.html)
- x-axis indicates the genre, y-axis shows the count for each genre.
- do not forget to put on the labels for the axes.

```python
# To-do
category = df['Category'].values

modes, counts = np.unique(category, return_counts=True)
```

```python
plt.bar(modes, counts)
plt.xticks(modes, rotation=90)
plt.xlabel('Category')
plt.ylabel('Count')
```

**(Optional)** You might have noticed that some primary categories are repeated. For instance, `Action` and `Action/Comedy` are listed as separate categories. Try searching online for a method to retain only the primary category in your dataframe. For example, `Action/Comedy` can be converted to `Action` since Action is the primary category. After cleaning up your dataset, try plotting the same graph again.

```python
# We create a function to cleanup our Category String

def clean_up_my_categories(category_name):
  return category_name.split('/')[0]
```

```python
# We are testing whether it is working as expected for all of our category names

for category_name in category:
    print(clean_up_my_categories(category_name))
```

```python
# Use pandas' apply method, and update the 'Category' column

df['Category'] = df['Category'].apply(clean_up_my_categories)
```

```python
# Plotting the updated values

category = df['Category'].values

modes, counts = np.unique(category, return_counts=True)

plt.bar(modes, counts)
plt.xticks(modes, rotation=90)
plt.xlabel('Category')
plt.ylabel('Count')
```

## Plot the histogram for the IMDB Ratings
- Use [plt.hist() function](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html)
- x-axis indicates the score interval, y-axis shows the number of movies within each score interval
- make the interval length 0.5
- do not forget to put on the labels for the axes.

- numpy arange() might come in handy!

```python
# To-do
min_score = np.min(rating)
max_score = np.max(rating)

bin_range = np.arange(min_score, max_score, 0.5)
plt.hist(rating, bin_range, ec='black')

plt.xlabel('Rating')
plt.ylabel('Number')
plt.show()
```

## Calculate the mean and variance of the IMDB Ratings of the movies


```python
# To-do
mean_score = np.mean(rating)
var_score = np.var(rating)

print("Mean Rating: ", mean_score)
print("Variance of Ratings: ", var_score)
```

**(Optional)** Try visualizing the Mean Rating on our histogram graph.

```python
# To-do
min_score = np.min(rating)
max_score = np.max(rating)

bin_range = np.arange(min_score, max_score, 0.5)
plt.hist(rating, bin_range, ec='black')
plt.axvline(mean_score, color='red', linestyle='dashed', linewidth=2, label="Mean Rating")

plt.xlabel('Rating')
plt.ylabel('Number')
plt.legend()
plt.show()
```

**(Optional)** You can also try to plot the covariance between IMDB ratings and coding experience

```python
# Extract the relevant columns
imdb_ratings = df["IMDB Rating"]
coding_experience = df["Coding experience (1-5)"]

# Calculate the covariance matrix
cov_matrix = np.cov(imdb_ratings, coding_experience)

# Extract the covariance value
covariance = cov_matrix[0, 1]

print(f"Covariance between IMDB Rating and Programming Experience: {covariance}")
```
